```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV files into DataFrames
df_orderlist = pd.read_csv('preprocessed_OrderList.csv')
df_freightrates = pd.read_csv('preprocessed_FreightRates.csv')
df_whcosts = pd.read_csv('preprocessed_WhCosts.csv')
df_whcapacities = pd.read_csv('preprocessed_WhCapacities.csv')
df_productsperplant = pd.read_csv('preprocessed_ProductsPerPlant.csv')
df_vmicustomers = pd.read_csv('preprocessed_VmiCustomers.csv')
df_plantports = pd.read_csv('preprocessed_PlantPorts.csv')
```

```python
print("WhCapacities Columns:", df_whcapacities.columns)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_orderlist['Weight'], bins=30, alpha=0.7, color='blue')
plt.title('Distribution of Order Weights')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_orderlist['Carrier'].value_counts().plot(kind='bar', color='orange')
plt.title('Number of Orders by Carrier')
plt.xlabel('Carrier')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Service Level', y='Unit quantity', data=df_orderlist)
plt.title('Unit Quantity by Service Level')
plt.xlabel('Service Level')
plt.ylabel('Unit Quantity')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_freightrates['rate'], bins=30, alpha=0.7, color='green')
plt.title('Distribution of Freight Rates')
plt.xlabel('Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_freightrates.groupby('Carrier')['minimum cost'].mean().sort_values().plot(kind='bar', color='purple')
plt.title('Average Minimum Cost by Carrier')
plt.xlabel('Carrier')
plt.ylabel('Average Minimum Cost')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_whcosts['Cost/unit'], bins=30, alpha=0.7, color='red')
plt.title('Distribution of Warehouse Costs per Unit')
plt.xlabel('Cost/unit')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_whcosts.set_index('WH')['Cost/unit'].plot(kind='bar', color='cyan')
plt.title('Cost per Unit by Warehouse')
plt.xlabel('Warehouse')
plt.ylabel('Cost/unit')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_whcapacities['Daily Capacity '], bins=30, alpha=0.7, color='magenta')
plt.title('Distribution of Daily Warehouse Capacities')
plt.xlabel('Daily Capacity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_whcapacities.set_index('Plant ID')['Daily Capacity '].plot(kind='bar', color='brown')
plt.title('Daily Capacity by Plant ID')
plt.xlabel('Plant ID')
plt.ylabel('Daily Capacity')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_productsperplant['Plant Code'].value_counts().plot(kind='bar', color='lime')
plt.title('Number of Products by Plant Code')
plt.xlabel('Plant Code')
plt.ylabel('Number of Products')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_vmicustomers['Plant Code'].value_counts().plot(kind='bar', color='teal')
plt.title('Number of Customers by Plant Code')
plt.xlabel('Plant Code')
plt.ylabel('Number of Customers')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_plantports['Plant Code'].value_counts().plot(kind='bar', color='pink')
plt.title('Number of Ports by Plant Code')
plt.xlabel('Plant Code')
plt.ylabel('Number of Ports')
plt.show()
```
